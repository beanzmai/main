```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
### Load Data from the Iris Dataset. 
### https://archive.ics.uci.edu/ml/datasets/iris

df = pd.read_csv(r'C:\Users\cg\Desktop\Echo\School\DSC609\Week 6\iris.csv')
```

```python
df.head() 
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
Y = df[['class']].values

from sklearn import preprocessing
X = preprocessing.scale(X)

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
Y = encoder.fit_transform(Y)

### Separate data into train/test datasets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

```python
### Next, we'll build our deep learning model. 
### Our specific neural network will have five layers, including the input and output layers.
### Each one of our layers will have a different activation function.
### We can also identify how many neurons should go into each layer

DeepNet = Sequential()

DeepNet.add(Dense(12, input_shape = (X.shape[1],), activation = 'relu', name = '1st'))
DeepNet.add(Dense(21, activation = 'relu', name = '2nd'))
DeepNet.add(Dense(15, activation = 'tanh', name = '3rd'))
DeepNet.add(Dense(8, activation = 'sigmoid', name = '4th'))
DeepNet.add(Dense(Y.shape[1], activation = 'softmax', name = 'output'))

optimizer = Adam(lr=0.001)
DeepNet.compile(optimizer, loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
### We'll print out the summary of our DeepNet model
### The summary provides us with the structure of our deep learning network.
### In addition, we can also see how many output nodes we have.

print(DeepNet.summary())
```

```python
### Now, we'll fit our model.
### We can see for each output ('Epoch') interation, we are provided an accuracy statstic.
### The accuracy statistic will increase as the number of interations increase.

DeepNet.fit(X_train,
            Y_train, 
            verbose = 1, 
            batch_size = X.shape[1],
            epochs = 215)
```

```python
DeepNet.evaluate(X_test, Y_test)
```
