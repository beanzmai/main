```python
import pandas as pd
import numpy as np
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn import model_selection
```

```python
### Load Data from the Early Stage Diabetes Risk Prediction Dataset. 
### This dataset contains the sign and symptpom data of newly diabetic or would be diabetic patient.
### https://archive.ics.uci.edu/ml/datasets/Early+stage+diabetes+risk+prediction+dataset.

df = pd.read_csv(r'C:\Users\cg\Desktop\Echo\School\DSC609\Week 5\diabetes.csv')
```

```python
### Check for null values
df.isnull().values.any()
```

```python
df.head() 
```

```python
### Data Preprocessing

### Rename the columns to replace the spaces with underscores.
df.columns = [c.replace(' ', '_') for c in df.columns]

### Replace Female/Male, No/Yes, Negative/Positive with 0/1.
df = df.rename(columns={'class': 'Class'})
df.Gender[df.Gender == 'Female'] = 0
df.Gender[df.Gender == 'Male'] = 1
df.Polyuria[df.Polyuria == 'No'] = 0
df.Polyuria[df.Polyuria == 'Yes'] = 1
df.Polydipsia[df.Polydipsia == 'No'] = 0
df.Polydipsia[df.Polydipsia == 'Yes'] = 1
df.sudden_weight_loss[df.sudden_weight_loss == 'No'] = 0
df.sudden_weight_loss[df.sudden_weight_loss == 'Yes'] = 1
df.weakness[df.weakness == 'No'] = 0
df.weakness[df.weakness == 'Yes'] = 1
df.Polyphagia[df.Polyphagia == 'No'] = 0
df.Polyphagia[df.Polyphagia == 'Yes'] = 1
df.Genital_thrush[df.Genital_thrush == 'No'] = 0
df.Genital_thrush[df.Genital_thrush == 'Yes'] = 1
df.visual_blurring[df.visual_blurring == 'No'] = 0
df.visual_blurring[df.visual_blurring == 'Yes'] = 1
df.Itching[df.Itching == 'No'] = 0
df.Itching[df.Itching == 'Yes'] = 1
df.Irritability[df.Irritability == 'No'] = 0
df.Irritability[df.Irritability == 'Yes'] = 1
df.delayed_healing[df.delayed_healing == 'No'] = 0
df.delayed_healing[df.delayed_healing == 'Yes'] = 1
df.partial_paresis[df.partial_paresis == 'No'] = 0
df.partial_paresis[df.partial_paresis == 'Yes'] = 1
df.muscle_stiffness[df.muscle_stiffness == 'No'] = 0
df.muscle_stiffness[df.muscle_stiffness == 'Yes'] = 1
df.Alopecia[df.Alopecia == 'No'] = 0
df.Alopecia[df.Alopecia == 'Yes'] = 1
df.Obesity[df.Obesity == 'No'] = 0
df.Obesity[df.Obesity == 'Yes'] = 1
df.Class[df.Class == 'Negative'] = 0
df.Class[df.Class == 'Positive'] = 1
```

```python
df.head() 
```

```python
### Assign variables

X = df[['Gender', 
        'Polyuria', 
        'Polydipsia', 
        'sudden_weight_loss', 
        'weakness', 
        'Polyphagia', 
        'Genital_thrush', 
        'visual_blurring', 
        'Itching', 
        'Irritability', 
        'delayed_healing', 
        'partial_paresis', 
        'muscle_stiffness', 
        'Alopecia', 
        'Obesity']].values

Y = df[['Class']].values.ravel()
Y = np.array(Y).astype(int)

from sklearn import preprocessing
X = preprocessing.scale(X)
```

```python
fold = model_selection.KFold(n_splits = 5, random_state = 2850)

### Import Multilayer Perceptron Classifier
from sklearn.neural_network import MLPClassifier

### Create a neural network (Single Hidden Layer with 100 neurons)
### Use Stochastic Gradient Descent (SGD) as the estimating technique. 
NN1 = MLPClassifier(hidden_layer_sizes = (100,), 
                    solver = 'sgd', 
                    learning_rate_init = .05, 
                    max_iter = 1000)

NN1_Accuracy = cross_val_score(NN1, X, Y, scoring = 'accuracy', cv = fold).mean()

NN1_Accuracy
```

```python
### Create a second neural network, this time with a hyperbolic tangent function activation function.

NN2 = MLPClassifier(hidden_layer_sizes = (100,), 
                    activation = 'tanh', 
                    learning_rate_init = .05, 
                    max_iter = 1000)

NN2_Accuracy = cross_val_score(NN2, X, Y, scoring = 'accuracy', cv = fold).mean()

NN2_Accuracy
```

```python
### Create a third neural network, this time with a logistic activation function.

NN3 = MLPClassifier(hidden_layer_sizes = (100,), 
                    activation = 'logistic', 
                    learning_rate_init = .05, 
                    max_iter = 1000)

NN3_Accuracy = cross_val_score(NN3, X, Y, scoring = 'accuracy', cv = fold).mean()

NN3_Accuracy
```

```python
### Interpret and Compare the results

### We created three different neural networks using various activation functions (hyperbolic tangent function & logistic) and estimating techniques (Stochastic Gradient Descent).
### All variables provided in the dataset were used when created the neural networks, which resulted in neural network models with fairly high acuracy between 94-95%.
### Certain estimating techniques/activation functions would be better suited for different types of data. 
### In our case and with our data, the most accurate model was the neural network which used Stochastic Gradient Descent (SGD) as the estimating technique.
```
